Finding Answers with the Data Using Visualizations

```python
# i.) What was the best year of sales?

df.groupby('year')['amount'].count().plot(kind='bar',title='Year Wise Sales')
```

```python
# ii.) Which was the best month for sales between 2015 to 2018

df_2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df_2015_2018.groupby('month')['rating'].count().plot(kind='bar')
```

```python
# iii.) What brand sold the most in the highest selling year(2015 to 2018)

df_2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df_2015_2018.groupby('brand')['amount'].sum().sort_values(ascending =False).head(10)\
.plot(kind='bar',title='Brand Wise Top 10 Sales 2015 to 2018',y='amount')
```

```python
# iv.) What products sold the most in the three years 2016, 2017 & 2018

#Creating subplots with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

#Plot for 2016
top_selling_2016 = df[df['year'] == 2016].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[0, 0].bar(top_selling_2016.index, top_selling_2016)
axs[0, 0].set_title('Top Selling Products in 2016')
axs[0, 0].tick_params(axis='x', rotation=45) 

#Plot for 2017
top_selling_2017 = df[df['year'] == 2017].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[0, 1].bar(top_selling_2017.index, top_selling_2017)
axs[0, 1].set_title('Top Selling Products in 2017')
axs[0, 1].tick_params(axis='x', rotation=45) 

#Plot for 2018
top_selling_2018 = df[df['year'] == 2018].groupby('brand')['rating'].count().sort_values(ascending=False).head(10)
axs[1, 0].bar(top_selling_2018.index, top_selling_2018)
axs[1, 0].set_title('Top Selling Products in 2018')
axs[1, 0].tick_params(axis='x', rotation=45)

#Hiding the empty subplot
axs[1, 1].axis('off')

#Adjusting layout for better appearance
plt.tight_layout()

#Showing the plots
plt.show()
```

```python
# v.) What product by category sold the most between 2015 to 2018?

df2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df2015_2018.groupby('category')['amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar',title='Top 10 Most Sold Product Category 2015 to 2018')
```

```python
# vi.)What product by category sold the least between 2015 to 2018?

df2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df2015_2018.groupby('category')['amount'].sum().sort_values(ascending=True).head(10).plot(kind='bar',title='10 Least Sold Product Brand 2015 to 2018')
```

```python
# vii.) What product by brand name sold the least between 2015 to 2018?

df2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df2015_2018.groupby('brand')['amount'].sum().sort_values(ascending= True).head(10).plot(kind='bar',title='10 Least Sold Product Brand 2015 to 2018')
```

```python
# viii.) Ratings Distribution

sns.countplot(x='rating', data=df)
```

```python
# ix.) Best rated brands

df2015_2018 = df[(df['year'] >= 2015) & (df['year'] <= 2018)]
df2015_2018.groupby('brand')['rating'].mean().sort_values(ascending =False).head(10).plot(kind='bar',title='10 most rating Brand 2015 to 2018')
```

```python
# x) Top 5 category sales percentage

df.groupby('category')['amount'].sum().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%',title='Top 5 category sales percentage')
```

```python
# xi) Brand wise sales percentage

df.groupby('brand')['rating'].count().sort_values(ascending=False).head(5).plot(kind='pie', autopct='%1.1f%%',title='Top 5 Brand wise sales percentage')
```

```python
# xii) Gender wise customer distribution

gender_distribution = df['gender'].value_counts()
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])
plt.title('Gender wise customer Distribution')
plt.show()
```
